```python
weights = []

weights += [25 * i for i in range(1, 2)]
weights += [50 * i for i in range(1, 20)]
weights += [200 * i for i in range(5, 10)]
weights += [1000 * i for i in range(2, 7)]

print(weights)
```

```python
import matplotlib.font_manager as fm
import os

# We assume that we have font file in the directory. 
# It is true for submission version,
# yet in general, we hold the font file in the repository root directory.
fontpath = "NanumGothic.ttf"
if not os.path.exists(fontpath):
    fontpath = "../" + fontpath

applyfont = "NanumGothic"
if not any(map(lambda ft: ft.name == applyfont, fm.fontManager.ttflist)):
    fm.fontManager.addfont(fontpath)
plt.rc("font", family=applyfont)
plt.rc("axes", unicode_minus=False)
```

```python
from math import ceil

def calculate_price(weight):
    if weight <= 50:
        return 520
    elif weight <= 1000:
        difference = weight - 50
        steps = ceil(difference / 50)
        base = calculate_price(50)
        return base + steps * 120
    elif weight <= 2000:
        difference = weight - 1000
        steps = ceil(difference / 200)
        base = calculate_price(1000)
        return base + steps * 120
    elif weight <= 6000:
        difference = weight - 2000
        steps = ceil(difference / 1000)
        base = calculate_price(2000)
        return base + steps * 400

results = list(map(calculate_price, weights))
print(weights)
print(results)
```

```python
import matplotlib.pyplot as plt

def annotate(xlist, ylist):
    if len(xlist) != len(ylist):
        return None
    for i in range(len(xlist)):
        plt.text(xlist[i], ylist[i], ylist[i], ha="center", fontsize=14)

plt.figure(figsize=(20,12))
plt.plot(weights[:-9], results[:-9], color='blue', drawstyle='steps-post', marker='o')
plt.yticks([100 * i for i in range(0, 31)])
annotate(weights[:-9], results[:-9])

plt.title(" (: 25g  1000g )", fontsize=30)
plt.xlabel("(: g)", fontsize=18)
plt.ylabel("(: )", fontsize=18)

plt.show()
```

```python

plt.figure(figsize=(20,12))
plt.plot(weights[-9:], results[-9:], color='blue', drawstyle='steps-post', marker='o')
annotate(weights[-9:], results[-9:])

plt.title(" (: 1kg  6kg )", fontsize=30)
plt.xlabel("(: g)", fontsize=18)
plt.ylabel("(: )", fontsize=18)
plt.yticks([100 * i for i in range(25, 51)])

plt.show()
```
